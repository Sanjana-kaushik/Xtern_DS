```python
#importing the required python libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#loading the dataset
df=pd.read_csv("2020-XTern-Ds.csv")
pd.DataFrame(df) #Using the python library panda to convert the data into a dataframe
```

```python
print(df.columns) #exploring the column names
print(df.head(5)) #Displaying the first five records
```

```python
df.dtypes #checking for datatypes
```

```python
df.duplicated().sum() #checking for duplicate values
```

```python
pd.DataFrame(round(df.isnull().sum()/df.shape[0] * 100,3), columns = ["Missing"]) #Looking for missing data just so that there will not be any issues when we work with data comparisions
```

```python
#Creating a count plot to display the various Ratings in the dataset
plt.rcParams['figure.figsize'] = 17,10
sns.countplot(df["Rating"], palette="Set1")
plt.title("Rating scenario")
plt.show()
```

```python
#Analyzing and displaying top ten cusines opted  
df.Cuisines.value_counts().nlargest(10).plot(kind = "barh")
plt.title("Top ten cuisines")
plt.xlabel("No of times chosen")
plt.show()

```

```python
#Showing trends in terms of Biggest restaurant chain 
plt.rcParams['figure.figsize'] = 14,7
plt.subplot(1,2,1)
df.Restaurant.value_counts().head().plot(kind = "barh", color = sns.color_palette("hls", 5))
plt.xlabel("No of Restaurants")
plt.title("Five biggest restaurant chains")

plt.subplot(1,2,2)
df[df['Rating']>='4.5']['Restaurant'].value_counts().nlargest(5).plot(kind = "barh", color = sns.color_palette("Paired"))
plt.xlabel("Rating")
plt.title("Top five restaurants based on Rating")
plt.tight_layout()
```

```python
#Analyzing the number of orders in the range of minimum orders
sns.countplot(df["Minimum_Order"], palette = "Set1")
plt.show()
```
